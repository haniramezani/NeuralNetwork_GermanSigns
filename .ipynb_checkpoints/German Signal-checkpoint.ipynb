```python

def pickle_original_imgs(X, labels, names, class_name, force):
    file_name=class_name+'.pickle'
    if os.path.exists(file_name) and not force:
        print('pickle file %s exists'%(file_name))
    else:
        data={'img':X, 'labels':labels, 'names':names}
        with open(file_name, 'wb') as handle:
            pickle.dump(data, handle, protocol=pickle.HIGHEST_PROTOCOL)
        print('pickeled %s'%(file_name))
    return 
```

```python
### this fucntion read pickled data and store matrixes in X and labels in labels. 
def read_pickles(path, num_pix, num_color):
    x=np.empty([0,num_pix, num_pix,num_color], dtype=np.uint8)
    l=np.empty([0])
    files = os.listdir(path)
    for file in files:
        file_type=file.split(".")[-1]
        if file_type=='pickle':
            with open(os.path.join(path,file), 'rb') as handle:
                d = pickle.load(handle)
            x=np.append(x,d['img'], axis=0 )
            l=np.append(l,d['labels'], axis=0 )
    return x, l
            
X,labels=read_pickles(r"C:\Users\STEVE\Neural Network-German Signals", 32, 3)
```

```python
###reads the first image of each class, these images will be used for visualization 
def get_class_sample(X, labels):
    sample_images=np.empty([0,32, 32,3], dtype=np.uint8)
    for i in np.unique(labels):
        ind=np.argwhere(labels==i).reshape(-1)[0]
        sample=X[ind,:,:,:].astype(np.uint8)
        sample_images=np.append(sample_images,[sample], axis=0 )
    return sample_images
sample_images=get_class_sample(X, labels)
```

```python
### display the sample of each class
def display_class_samples(class_samples, cols):
    fig = plt.figure(figsize=(15,10))
    n_image=len(class_samples)
    titles = ['Class (%d)' % i for i in range(0,n_image )]
    for n, (sample, title) in enumerate(zip(sample_images, titles)):
        a = fig.add_subplot(cols, np.ceil(n_image/float(cols)), n + 1)
        plt.imshow(sample)
        a.set_title(title)
    plt.tight_layout()
    plt.show()
display_class_samples(sample_images, 5)
```

```python

## this fucntion determine the frequency of each class. it is used for the original data set and later it will be used for augmented data set
def get_frequency(labels):
    num_class=43
    count=[0]*num_class
    for i in range(num_class):
        size=labels[labels==i].shape[0]
        count[i]=size
        
    return np.array(count)
#### plot the frequency which is the output of "def get_frequency"
def plot_frequesncy(labels_index, lables_count):
    w=0.4
    fig, ax1=plt.subplots(1,1, figsize=(10,6))
    p1=ax1.bar(labels_index+w/2, c,  width=w, color='darkBlue')
    ax1.axhline(y=2000, ls='--')
    ax1.set_xlabel('Class index')
    ax1.set_ylabel('Number of images')
    plt.show()

c=get_frequency(labels)       
plot_frequesncy(np.unique(labels), c)   
```

```python
#### this is to set the threshold above which we do not need to augment data. Based on the plot above, I use 2000
### during augmentation, this fuction is called for every class. if it returns true, then I do not do augmentation
def Sampleisenough(count_array, classindex):
    min_sample_size=2000
    return count_array[int(classindex)]>min_sample_size
```

```python
### this fucntion change the matrixes according to the type of mirror. there are three types of mirror:
### 1- horizontal denoted by 'h' 2-vertical: 'v' and 3- both vertical and horizontal 'hv'
def mirror_single_class(M,l, mirror_type):
    if mirror_type=='h':  return M[:, :, ::-1, :], np.repeat(l,M.shape[0])
    elif mirror_type=='v': return M[:, ::-1, :, :], np.repeat(l,M.shape[0])
    elif mirror_type=='hv': return M[:, ::-1, ::-1, :], np.repeat(l,M.shape[0])
    else: 
        Print("inappropriate mirror_type. It has to be either 'h', 'v', or 'hv ")
        return 

#### i do mirroring for a batch of data which includes multiple classes
def mirror_multiple_classes(X,labels, class_list, mirror_type):
    X_=np.empty([0,32, 32,3], dtype=np.float32)
    labels_=np.empty([0])
    
    if mirror_type!='xh':
        for i in class_list:
            if Sampleisenough(c, i): continue
            mirrored, mirror_label=mirror_single_class(X[labels==i],i, mirror_type)
            X_=np.append(X_,mirrored, axis=0)
            labels_=np.append(labels_,mirror_label, axis=0)
    else: 
        ### this is to implement the cross horizontal mirror when the class of data changes after mirroring.
        ### it is espetial case of horizontal mirror
        for pair in class_list:
            if not Sampleisenough(c, pair[1]):
                mirrored, mirror_label=mirror_single_class(X[labels==pair[0]],pair[1], 'h')
                X_=np.append(X_,mirrored, axis=0)
                labels_=np.append(labels_,mirror_label, axis=0)
            if not Sampleisenough(c, pair[1]):
                xmirrored, xmirror_label=mirror_single_class(X[labels==pair[1]],pair[0], 'h')
                X_=np.append(X_,xmirrored, axis=0)
                labels_=np.append(labels_,xmirror_label, axis=0)
            
    return X_, labels_


```

```python
###### this lists determine the classes for each type of mirroring
h_symmetric = [11, 12, 13, 15, 17, 18, 22, 26, 30, 35]
v_symmetric = [1, 5, 12, 15, 17]
hv_symmetric = [32, 40]
### I will do mirroring for both first and second element of each list item in the following list
x_symmetric = [[19, 20], [33, 34], [36, 37], [38, 39] ]

### this is to call the fuction to mirror data
x_mirrored,l_mirrored=[],[]
for class_list, mirror_type in zip([h_symmetric, v_symmetric, hv_symmetric, x_symmetric],['h','v','hv','xh']):
    out=mirror_multiple_classes(X,labels, class_list, mirror_type)
    x_mirrored.append(out[0])
    l_mirrored.append(out[1])
```

```python
### append mirrored data to the original data
for a,b in zip(x_mirrored, l_mirrored):
    X=np.append(X, a, axis=0)
    labels=np.append(labels, b, axis=0)
#### delete variables to release memory
del x_mirrored, l_mirrored    
    
```

```python
##### this fucntion rotate a single image
def rotate_img(Images, angle):
    Images_rotate=np.empty([0,Images.shape[1], Images.shape[2],Images.shape[3]], dtype=np.uint8)
    for n, img in enumerate(Images):
        ang=np.random.uniform(-angle, angle,size=1)
        img_rotate = rotate(img,ang )
        Images_rotate=np.append(Images_rotate,[img_rotate], axis=0)
    return Images_rotate 

```

```python
#### this rotates a bach of images
def rotate_batch(X, labels, note, Freq, degree):
    x_=np.empty([0,32,32,3], dtype= np.int8)
    l_=np.empty([0], dtype= np.int8)
    for class_index in np.arange(43):
        
        if Sampleisenough(Freq, int(class_index)): 
            continue
        else:
            x=rotate_img(X[labels==class_index],degree)
            l=labels[labels==class_index]
            x_=(np.append(x_,x, axis=0))
            l_=(np.append(l_,l, axis=0))
            print(class_index,end=" ")
    
    #pickle_imgs2(x_, l_, r"C:\Users\STEVE\Neural Network-German Signals", 'rotate_batch',note,False)
    return x_, l_
```

```python
#############this is to transform a single image
def projection_transform(X, intensity):
    
    image_size = X.shape[1]
    d = image_size * 0.3 * intensity
    for x in X:
        tl, bl, tr, br=([] for i in range(4))
        corner_shifts=[tl, bl, tr, br] ### represent four corners of an immage
        for corner_shift in corner_shifts:
            corner_shift.append(np.random.uniform(-d, d, size=1)) #####this is for x coordinate
            corner_shift.append(np.random.uniform(-d, d, size=1)) ####this is for y coordinate

        shift=np.array([tl,bl,tr,br]).reshape(4,2)
        destination=np.array([[0, 0],[0, image_size],[image_size, image_size], [image_size, 0]]).reshape(4,2)
        source=destination-shift
        
        transform = ProjectiveTransform()
        transform.estimate(source,destination)
        x = warp(x, transform, output_shape=(image_size, image_size), order = 1, mode = 'edge')
        
    return X
```

```python
#### this is to transform a batch of data
def transform_batch(X, labels, note, Freq):
    x_=np.empty([0,32,32,3], dtype= np.int8)
    l_=np.empty([0], dtype= np.int8)
    for class_index in np.arange(43):
        
        if Sampleisenough(Freq, int(class_index)): 
            continue
        else:
            x=projection_transform(X[labels==class_index], 0.75)
            l=labels[labels==class_index]
            x_=(np.append(x_,x, axis=0))
            l_=(np.append(l_,l, axis=0))
            print(class_index,end=" ")
    
    #pickle_imgs2(x_, l_, r"C:\Users\STEVE\Neural Network-German Signals", 'rotate_batch',note,False)
    return x_, l_
```
